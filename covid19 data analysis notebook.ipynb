### Task 2.2: Delete the useless columns

same with columns

```python
dataset_covid19.drop(["Lat","Long"],axis=1)
```

```python
dataset_covid19.drop(["Lat","Long"],axis=1, inplace=True)
dataset_covid19.head()
```

### Task 2.3: Aggregating the rows by the country
groupby groups based on countries, sum adds up all the content columnwise where country is same

```python
countrywise_covid19 = dataset_covid19.groupby("Country/Region").sum()
```

```python
countrywise_covid19.head()
```

```python
countrywise_covid19.shape
```

### Task 2.4: Visualizing data related to a country for example China

### Task3: Calculating a good measure

### task 3.2: find maxmimum infection rate for China

```python
countrywise_covid19.loc['India']["3/22/20":].diff().max()
```

```python
countrywise_covid19.loc['India']["3/22/20":].max()
```

```python
countries= list(countrywise_covid19.index)
print(countries)
max_rates=[]
for c in countries:
    max_rates.append(countrywise_covid19.loc[c].diff().max())
print(max_rates)
```
